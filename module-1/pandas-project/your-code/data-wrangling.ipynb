### 1. Import pandas library

```python
import pandas as pd
```

### 2. Import the data using Pandas

```python
data = pd.read_csv("GSAF5.csv", encoding = "ISO-8859-1")
data.head()
data.dtypes
```

### 3. Examining Data for Potential Issues
#### Missing values

```python
null_cols = data.isnull().sum()
null_cols[null_cols > 0]
```

#### Drop empty columns

```python
data = data.drop('Unnamed: 22', axis=1)
data = data.drop('Unnamed: 23', axis=1)
data
```

#### Finding and Removing Duplicates

#####  Remove duplicated columns href and href_formula

```python
data = data.drop('href formula', axis=1)
data
```

##### Remove duplicate records

```python
before = len(data)
select_columns = ['Date', 'Year', 'Type', 'Country', 'Area', 'Location']
data = data[select_columns].drop_duplicates()
after = len(data)
print('Number of duplicate records dropped: ', str(before - after))
```

#### Changing the ages with not numeric values by cero

```python
data['Age'] = data['Age'].str.extract("([+]?\d*\.\d+|[+]?\d+)").astype(float)
data['Age'] = data['Age'].fillna(value=0)
data
```

#### Data Type Correction

```python
data['Year'] = data['Year'].astype('object')
data['Year'].dtype
```

#### Cleaning Text and Removing Special Characters

##### Cleaning "Date" column which has the words "Reported", "No date, Before", "Late" in some rows.

```python
data['Date'] = data['Date'].str.replace('Reported', '')
data['Date'] = data['Date'].str.replace('No date, Before', '')
data['Date'] = data['Date'].str.replace('Late', '')
print(set(data['Date']))
```
