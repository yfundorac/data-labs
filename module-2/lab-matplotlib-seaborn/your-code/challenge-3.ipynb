#### Whether the bin size affects your plot? Why?

```python
#your comment here
'''The bin width (and thus number of categories or ranges) affects the ability of a histogram to identify local 
regions of higher incidence. Too large, and you will not get enough differentiation. Too small, and the data cannot
be grouped.

Explanation:
A good histogram will show areas of higher incidence of a parameter that may help us to identify causative factors
in a system. Thus, inappropriate bin size will defeat the purpose of the histogram. The extremes may help visualize
the effect. ONE 'bin' only shows the population or total sample size. A 'bin' for each sample point gives us no more
information, but only stretches the width of the chart.

A good bin width will usually show a recognizable normal probability distribution curve, unless the data is really
multi-modal. Then there could be two or more distinct 'humps' in the histogram chart.'''
```
